```python
import pickle

# Wczytaj feature_names
with open('../models/feature_names.pkl', 'rb') as f:
    feature_names = pickle.load(f)

# Pokaż wszystkie kolumny condition
condition_cols = [col for col in feature_names if col.startswith('condition_')]
print("="*60)
print("KOLUMNY CONDITION W MODELU:")
print("="*60)
for i, col in enumerate(condition_cols, 1):
    print(f"{i}. {col}")

# Pokaż wszystkie kolumny model
model_cols = [col for col in feature_names if col.startswith('model_')]
print("\n" + "="*60)
print("KOLUMNY MODEL W MODELU:")
print("="*60)
for i, col in enumerate(model_cols, 1):
    print(f"{i}. {col}")

# Pokaż WSZYSTKIE feature names (pierwszy 10)
print("\n" + "="*60)
print("WSZYSTKIE FEATURES (pierwsze 10):")
print("="*60)
for i, col in enumerate(feature_names[:10], 1):
    print(f"{i}. {col}")

print(f"\nCałkowita liczba features: {len(feature_names)}")
```
